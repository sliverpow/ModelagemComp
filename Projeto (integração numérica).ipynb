```python
import math
from matplotlib import pyplot as plt
```

```python
#incialização dos valores inciais do problema 
theta= math.pi/3
w=0
h=0.1
t=0
g=9.8
l=9.8
i=0
angulo=[]
tempo=[]
omega=[]
```

```python
#O método euler-chromer foi utilizado para a o calculo de angulo e posição, este método funciona muito bem para angulos pequenos,
#onde a aproximação de sen(theta) converge.
while (t<10000):
  w=w-theta*h*g/l
  theta=theta+w*h
  i=i+1
  angulo.append(theta)
  t=t+h*i
  tempo.append(t)
  omega.append(w)
```

```python
#Os dois graficos abaixo mostram a variação oscilatória do angulo e velocidade angular do Pendulo
# o primeiro gráfico é referente ao omega(velocidade angular)
# o segundo é refente ao angulo theta formado pela aste do pendulo com a superficie onde esta ligado
plt.plot(tempo,omega)
plt.show()
plt.plot(tempo,angulo)
plt.show()
```

```python
#Parte 2 Solução numérica
#Definição dos valores iniciais
a=0.1
h=1
P=100
tfinal=100
t=0
tempo=[]
populacao=[]
```

```python
while (t<tfinal):
  P=P+a*P*h
  t=t+h
  populacao.append(P)
  tempo.append(t)
```

```python
#Parte 2 solução Analítica
t=0
P=100
Panalitico=[]
while (t<tfinal):
  P2=P*math.exp(a*t)
  t=t+h
  Panalitico.append(P2)
```

```python
# O primeiro gráfico é referente ao resultado obtido pela integração numérica da função
plt.plot(tempo,populacao)
plt.show()
# O segundo é referente ao valor análito da função
plt.plot(tempo,Panalitico)
plt.show()
# Como observado eles apresentam valores muito similares, e o mesmo comportamento. Isso prova que esse método é eficaz para essa faixa de valores.
# O que podemos argumentar sobre a função em sí, é que ela não possui um controle, um limite, e para um caso de população isso seria inviável, pois não haveria alimentos suficientes.
```

# PARTE 1 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#Importando os vetores com as posições
x=np.array([-0.47, -0.26, 0.15, 0.82, -0.60])
y=np.array([1.14,1.21,1.28,1.47,0.93])

#Criando a Matrix A
A=np.column_stack([x**0,x])

m = np.dot(A.T, A)
c = np.dot(A.T, y)

#realindo a equação (20)
i1, slope1 = np.dot(np.linalg.inv(m), c)
print(i1, slope1)

#plotando o gráfico
plt.plot(x, y, 'bo')
plt.plot(x, np.dot(A, [i1, slope1]), 'r--')
plt.xlabel('x')
plt.ylabel('y')

```

# Parte 2

```python
#Importando os vetores com as posições
x=np.array([-0.47, -0.26, 0.15, 0.82, -0.60])
y=np.array([0.12,0.25,0.18,0.26,-0.11])

#Criando a Matrix A
A=np.column_stack([x**0,x,x**2,x**3])


m = np.dot(A.T, A)
c = np.dot(A.T, y)

#realindo a equação (20)
a0,a1,a2,a3 = np.dot(np.linalg.inv(m), c)
print(a0,a1,a2,a3)

#usando os paramentros para plotar o gráfico 
y2=[]
x2=[]

for i in x:
  x2=np.sort(np.append(x2,i))
  y2=np.sort(np.append(y2,(a0+a1*i+a2*i**2+a3*i**3)))


#plotando o gráfico
plt.plot(x, y, 'bo')
plt.plot(x2,y2 , 'r--')
plt.xlabel('x')
plt.ylabel('y')
```

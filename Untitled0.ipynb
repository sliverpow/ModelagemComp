```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5,6,1)
y = np.arange(-5,6,1)

X, Y = np.meshgrid(x, y)

u, v = -X, -Y

fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(X,Y,u,v)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([-6, 6, -6, 6])
ax.set_aspect('equal')

plt.show()
```

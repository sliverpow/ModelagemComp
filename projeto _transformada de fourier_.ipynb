```python
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
import numpy as np
import math
import time

#contrução das "funções"
N1= 100
t=1/N1

y1=np.zeros(N1)
for a in range(1,N1):
  y1[a]=math.sin(2*np.pi*t*a)

y2=np.zeros(N1)
for a in range(1,N1):
  y2[a]=1+ 2*math.cos(2*np.pi*t*a)

y3=np.zeros(N1)
for a in range(1,N1):
  y3[a]=1+ 2*(math.cos(2*np.pi*t*a)+math.cos(2*2*np.pi*t*a)+math.cos(2*3*np.pi*t*a))

y4=np.zeros(N1)
for a in range(1,N1):
  y4[a]=1+2*(math.cos(2*np.pi*t*a)+math.cos(2*2*np.pi*t*a)+math.cos(2*3*np.pi*t*a)+math.cos(2*4*np.pi*t*a)+math.cos(2*5*np.pi*t*a))

y5=np.zeros(N1)
for a in range(1,N1):
  y5[a]=math.cos(2*np.pi*t*a)-math.cos(2*np.pi*t*a)*1/3

y6=np.zeros(N1)
for a in range(1,N1):
  y6[a]=math.cos(2*np.pi*t*a)-math.cos(2*np.pi*t*a)*1/3+math.cos(10*np.pi*t*a)*1/5

y7=np.zeros(N1)
for a in range(1,N1):
  y7[a]=math.cos(2*np.pi*t*a)-math.cos(2*np.pi*t*a)*1/3+math.cos(10*np.pi*t*a)*1/5+math.cos(14*np.pi*t*a)*1/7

y8=np.zeros(N1)
for a in range(1,N1):
  y8[a]=100

y9=np.zeros(N1)
for a in range(1,N1):
  y9[a]=math.cos(2*np.pi*t*a)


```

```python
#Construção da DFT

def dft(y,N):
  c=np.zeros(N,complex)
  n=np.arange(N)
  for k in range(N):
    c[k]=np.sum(y*np.exp(-2j*np.pi*k*n/N))

  plt.figure(figsize=(12,4))
  plt.subplot(1,2,1)
  plt.plot(np.imag(c), label="img")
  plt.plot(np.real(c), label="real")
  plt.legend()


  plt.show()

#Construção da fft

def afft(y,N):

  c=np.zeros(N,complex)
  c = fft(y)

  plt.figure(figsize=(12,4))
  plt.subplot(1,2,1)
  plt.plot(np.imag(c), label="img")
  plt.plot(np.real(c), label="real")
  plt.legend()
  plt.show()

```

## Comparando os tempos de calculo da fft e da dft para função seno, cosseno, soma de cossenos, e onda quadrada.

```python
#Seno
print("dft","seno")
start_time = time.time()
dft(y1,N1)
print(" tempo de execução dft",(time.time() - start_time))
print("fft","seno")
start_time = time.time()
afft(y1,N1)
print(" tempo de execução fft",(time.time() - start_time))

print("dft","cosseno")
start_time = time.time()
dft(y9,N1)
print(" tempo de execução dft",(time.time() - start_time))
print("fft","Cosseno")
start_time = time.time()
afft(y9,N1)
print(" tempo de execução fft",(time.time() - start_time))

print("dft","Somatório de cossenos")
start_time = time.time()
dft(y3,N1)
print(" tempo de execução dft",(time.time() - start_time))
print("fft","Somatório de cossenos")
start_time = time.time()
afft(y3,N1)
print(" tempo de execução fft",(time.time() - start_time))

print("dft","Onda Quadrada")
start_time = time.time()
dft(y5,N1)
print(" tempo de execução dft",(time.time() - start_time))
print("fft","Onda Quadrada")
start_time = time.time()
afft(y5,N1)
print(" tempo de execução fft",(time.time() - start_time))



```
